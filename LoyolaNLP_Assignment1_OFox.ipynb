# 1. [20 pts] 
Define each of the following machine learning terms:
    - dataset
    - training, testing, validation dataset
    - ground truth, label
    - pre-processing, feature, numerical, nominal
    - decision surface
    - model validation, accuracy, cross-validation

Dataset: Preferably untouched raw information (images, csv, etc)

Training, Testing, Validation Dataset: The training dataset is the portion used to train a model, while the testing dataset is the non-overlapping portion used to test the trained model. A validation data set is reserved for the final tuned model, and contains data that the model has not seen before (has neither been trained or tested with).

Ground Truth, Label: This is the known correct result for the data field that the model is being trained to predict.

Pre-processing, Feature, Numerical, Nominal: Prep-processing is the set of steps needed to prune erroneous data points (duplicates, outliers, etc) and put it into a format that a ML model will accept (flattening, etc). A feature is a specific field that each data point has (Person might be the data point, and Age, Height might be the features). Numerical data consist of Integers and Floats, used in fields that are represented with numbers. Nominal data are essentially anything other than numerical; they can be True/False (boolean), or strings such as the title of a company, a street name, etc.

Decision Surface: A decision surface is the divider for a classifier, this can be a line for 2-D data, a plane for 3-D, or a hyperplane for n-D data.

Model Validation, Accuracy, Cross-validation: Model validation consists of testing a model with real world data to see if the models predictions are realistic. Model accuracy is calculated by ( TP + TN ) / ( TP + FP + TN + FN ). Cross validation consists of partitioning a set of data into multiple parts, then training on one part, and testing on another, and rotating through the partitions, averaging the validation/test results to estimate predictive performance.

Pearson Correlation coefficient: 

r = sum( x - m_x )( y - m_y ) / sqrt( sum( x - m_x )^2 sum( y - m_y )^2  )

```python
admin_df2.head( )
```

```python
admin_corr = pd.DataFrame( columns = admin_df2.columns, index = admin_df2.columns )

for x in range( admin_df2.shape[ 1 ] ):
    
    col_x = admin_df2.columns[ x ]
    
    m_x = np.mean( admin_df2[ '%s' % col_x ] )
    
    for y in range( admin_df2.shape[ 1 ] ):
        
        # Sums
        x_diff, x_diff_sq, y_diff, y_diff_sq = 0, 0, 0, 0
        num_prod = 0
        
        col_y = admin_df2.columns[ y ]
    
        m_y = np.mean( admin_df2[ '%s' % col_y ] )
        
        for row in range( admin_df2.shape[ 0 ] ):
        
            x_0 = admin_df2[ '%s' % col_x ][ row ]
            
            y_0 = admin_df2[ '%s' % col_y ][ row ]
            
            x_diff += x_0 - m_x
            x_diff_sq += ( x_0 - m_x ) ** 2
            
            y_diff += y_0 - m_y
            y_diff_sq += ( y_0 - m_y ) ** 2
            
            num_prod += ( x_0 - m_x ) * ( y_0 - m_y )
            
        PearsonCoeff = num_prod / np.sqrt( x_diff_sq * y_diff_sq  )
        
        admin_corr.loc[ [ '%s' % col_y ], [ '%s' % col_x ] ] = PearsonCoeff
```

```python
def encode_onehot(_df, f):
    _df2 = pd.get_dummies(_df[f], prefix='', prefix_sep='').max(level=0, axis=1).add_prefix(f+' - ')
    df3 = pd.concat([_df, _df2], axis=1)
    df3 = df3.drop([f], axis=1)
    return df3
```

```python
shroom_df2 = encode_onehot( shroom_df, 'cap-shape' )
```

```python
for i in range( 2, len( shroom_df.columns ) ):
    shroom_df2 = encode_onehot( shroom_df2, shroom_df.columns[ i ] )
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import KFold, StratifiedKFold, train_test_split
```

```python
def rf_train_test(_X_tr, _X_ts, _y_tr, _y_ts):
    # Create a new random forest classifier, with working 4 parallel cores
    rf = RandomForestClassifier(n_estimators=200, max_depth=5, random_state=None, n_jobs=4)
    # Train on training data
    model = rf.fit(_X_tr, _y_tr)
    # Test on training data
    y_pred = rf.predict(_X_ts)
    # Return accuracy
    return accuracy_score(_y_ts, y_pred)
```

```python
X = shroom_df2.loc[:, shroom_df2.columns != 'class'].values
y = shroom_df2.loc[:, shroom_df2.columns == 'class'].values.ravel()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=None)
rf_train_test(X_train, X_test, y_train, y_test)
```

```python
for i in range(10):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=None)
    print(rf_train_test(X_train, X_test, y_train, y_test))
```

```python
accuracies = []
for i in range(100):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=None)
    accuracies += [rf_train_test(X_train, X_test, y_train, y_test)]

print(f'80% train-test split accuracy is {np.mean(accuracies):.3f} {chr(177)}{np.std(accuracies):.4f}')
```

```python
accuracies = []
kf = KFold(n_splits=10,shuffle=False,random_state=None)
for train_index, test_index in kf.split(X, y):
    acc = rf_train_test(X[train_index], X[test_index], y[train_index], y[test_index])
    accuracies += [acc]

print(f'10-fold cross validation accuracy is {np.mean(accuracies):.3f} {chr(177)}{np.std(accuracies):.4f}')
```
